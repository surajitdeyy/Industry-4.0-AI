```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
```

```python
csv_data = pd.read_csv("/home/ubuntu/Data/AI4AOI/train/train.csv")
```

```python
train_path = "/home/ubuntu/Data/AI4AOI/train/"
dest_train_path = "/home/ubuntu/Data/AOI_training/train/"
```

```python
bitmap_list = ["bitmap_files_0", "bitmap_files_1" ,"bitmap_files_2" ,"bitmap_files_3" ,"bitmap_files_4" ,"bitmap_files_5" ,"bitmap_files_6" ,"bitmap_files_7" ,"bitmap_files_8"]
```

```python
for index, row in csv_data.iterrows():
    print("Processing {} image..".format(index+1))
    count = 0
    list_nan = row.isna()[row.isna() == True].index.get_values()
    missing_images = len(list_nan)
    assert(os.path.exists(os.path.join(train_path, str(row['json_content.component.verification code']),  row['parent_name'], row['png_file'])))
    grayscale_image = cv2.imread(os.path.join(train_path, str(row['json_content.component.verification code']),  row['parent_name'], row['png_file']))
    grayscale_image = cv2.cvtColor(grayscale_image, cv2.COLOR_BGR2GRAY)
    grayscale_image = cv2.resize(grayscale_image, (256, 256)) 
    temp_image = None
    
    for bitmap in bitmap_list:
        if bitmap not in list_nan and np.random.randint(0,2)==1:
            opt2gray_image = cv2.imread(os.path.join(train_path,  str(row['json_content.component.verification code']), row['parent_name'], row[bitmap]))
            opt2gray_image = cv2.cvtColor(opt2gray_image, cv2.COLOR_BGR2GRAY)
            opt2gray_image = cv2.resize(opt2gray_image, (256, 256)) 

            if temp_image is None:
                if count <= 2:
                    temp_image = cv2.merge((grayscale_image, opt2gray_image))
                    count += 1
            else:
                if count <= 2:
                    temp_image = cv2.merge((temp_image, opt2gray_image))
                    count += 1
    no_of_layers = 3 - count
    print(temp_image.shape, grayscale_image.shape)
    for i in range(no_of_layers ):
        if temp_image is None:
            temp_image = cv2.merge((grayscale_image, grayscale_image))
    assert(temp_image.shape[2] == 4)
    
    if(row['json_content.component.repair result']==1):
        dest_path = os.path.join(dest_train_path, "Bad_parts")
    else:
        dest_path = os.path.join(dest_train_path, "Good_parts")
    cv2.imwrite(os.path.join(dest_path, row['png_file']), temp_image)
```
